```python
# Find overlapping columns between the two DataFrames
overlapping_columns = merged_df.columns.intersection(labs_df.columns)

# Define a function to compare and merge values in overlapping columns
def compare_and_merge(df1, df2, column, tolerance=1e-9):
    discrepancies = []
    for index in df1.index.intersection(df2.index):
        value1 = df1.at[index, column]
        value2 = df2.at[index, column]
        
        # Convert both values to floats if they are not NaN
        if not pd.isna(value1):
            value1 = float(value1)
        if not pd.isna(value2):
            value2 = float(value2)
        
        # Check if both values are the same within the specified tolerance
        if pd.isna(value1) and not pd.isna(value2):
            df1.at[index, column] = value2
        elif not pd.isna(value1) and not pd.isna(value2):
            if not abs(value1 - value2) <= tolerance:
                # Log discrepancies instead of raising an error
                discrepancies.append((index, column, value1, value2))
    
    # Handle discrepancies as needed, e.g., print them, log to a file, etc.
    for discrepancy in discrepancies:
        print(f"Discrepancy at index {discrepancy[0]} for column {discrepancy[1]}: {discrepancy[2]} vs {discrepancy[3]}")
    
    return df1

# Process only the overlapping columns
for column in overlapping_columns:
    merged_df = compare_and_merge(merged_df, labs_df, column)

# Now, concatenate the non-overlapping columns from labs_df to merged_df
non_overlapping_columns = labs_df.columns.difference(merged_df.columns)
merged_df = pd.concat([merged_df, labs_df[non_overlapping_columns]], axis=1)
```

```python
import numpy as np

def merge_columns(series_1, series_2):
    series_1, series_2 = series_1.align(series_2, fill_value=np.nan)
    merged_values = np.where(series_1.isnull(), series_2, 
                    np.where(series_2.isnull(), series_1, 
                    series_1))
    flags = np.where((~series_1.isnull() & ~series_2.isnull() & (series_1 != series_2)), "FLAG", None)
    return merged_values, flags
```

```python
def process_overlaps(df1, df2, overlapping_columns):
    # Making a copy of the first dataframe to store results
    result_df = df1.copy()
    
    for col in overlapping_columns:
        # Join on the index (PERSONID)
        temp_df = df1[[col]].join(df2[[col]], rsuffix="_r", how='left')
        
        # Logic to create merged values
        # 1) if both columns have same value for one patient, then we can keep the value
        # 2) if there is value in one column but not other, we take the value one
        # 3) if there is value mismatch, we flag it
        mask_same = temp_df[col] == temp_df[col + "_r"]
        mask_one_missing = temp_df[col].isnull() | temp_df[col + "_r"].isnull()
        mask_mismatch = ~(mask_same | mask_one_missing)
        
        result_df[col + "_merged"] = np.where(mask_same, temp_df[col], 
                                  np.where(mask_one_missing, temp_df[col].fillna(temp_df[col + "_r"]), 
                                  "FLAGGED"))
        
        # If you want to retain the values when flagged:
        result_df[col + "_merged"] = np.where(mask_mismatch, 
                                  temp_df[col].astype(str) + "|" + temp_df[col + "_r"].astype(str),
                                  result_df[col + "_merged"])
        
        # Drop the original column
        result_df = result_df.drop(col, axis=1)
        
        # Rename the merged column to original name
        result_df = result_df.rename(columns={col + "_merged": col})
        
    return result_df
```

```python
def merge_columns(row, col1, col2):
    val1 = row[col1]
    val2 = row[col2]
    
    if pd.isnull(val1):
        return val2
    if pd.isnull(val2):
        return val1
    if val1 == val2:
        return val1
    return "FLAG"

combinations = [
    (clinical_df, labs_df, 'clinical_df', 'labs_df'),
    (clinical_df, drugs_df, 'clinical_df', 'drugs_df'),
    (labs_df, drugs_df, 'labs_df', 'drugs_df')
]

for df1, df2, df1_name, df2_name in combinations:
    overlapping_cols = df1.columns.intersection(df2.columns).tolist()
    overlapping_cols = [col for col in overlapping_cols if col not in ['POMPE']]
    
    for col in overlapping_cols:
        combined_col_name = f"{col}_combined"
        
        # This step creates two new dataframes with the same index but different column names for merging
        merged_df = df1[[col]].rename(columns={col: f"{col}_{df1_name}"}).join(df2[[col]].rename(columns={col: f"{col}_{df2_name}"}))
        merged_df[combined_col_name] = merged_df.apply(merge_columns, col1=f"{col}_{df1_name}", col2=f"{col}_{df2_name}", axis=1)
        
        # Update original dataframes with the new combined column and drop old columns
        df1.drop(columns=col, inplace=True)
        df1[combined_col_name] = merged_df[combined_col_name]
        
        df2.drop(columns=col, inplace=True)
        df2[combined_col_name] = merged_df[combined_col_name]
```

```python
def get_flagged_rows(df):
    return df[df.applymap(lambda x: x == "FLAG").any(axis=1)]

# Now, for each dataframe, fetch the flagged rows
flagged_rows_clinical = get_flagged_rows(clinical_df)
flagged_rows_labs = get_flagged_rows(labs_df)
flagged_rows_drugs = get_flagged_rows(drugs_df)

# To see the flagged rows for each dataframe:
print("Flagged rows in clinical_df:")
print(flagged_rows_clinical)

print("\nFlagged rows in labs_df:")
print(flagged_rows_labs)

print("\nFlagged rows in drugs_df:")
print(flagged_rows_drugs)
```
